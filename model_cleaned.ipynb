# Model

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, test_size=0.1)
train_X, train_y = train['text'], train['label']
test_X, test_y = test['text'], test['label']
```

```python
import tensorflow as tf
tf.test.is_gpu_available()
```

```python
max_seq_len = data['text'].apply(lambda t: len(t.split())).max()
max_seq_len

data['text'].apply(lambda t: len(t.split())).nlargest()
```

```python
from word2vec_keras import Word2VecKeras
model = Word2VecKeras()
model.train(train_X.to_list(),
            train_y,
            w2v_size=50,  # much smaller n words
            w2v_min_count=1,
            w2v_epochs=100,
            k_batch_size=32,
            k_epochs=300,
            k_max_sequence_len=max_seq_len,
            k_lstm_neurons=10)
```

```python
res = []

for text, label in zip(test_X, test_y):
    try:
        pred = model.predict(text)
        res.append(pred)
        print(f"{pred}, orig: {label}")
    except IndexError as e:
        print(e)
        res.append(None)
```

```python
from sklearn.metrics import accuracy_score

none_count = 0
res2 = []
test2 = []

for r, t in zip(res, test_y):
    if r is not None:
        res2.append(r['label'])
        test2.append(t)
    else:
        none_count += 1
        
print(none_count)
accuracy_score(test2, res2)
```

```python
import numpy as np

def dummy_val(r):
    return r['label'] if r is not None else -1

test2_arr = np.array(test_y)
res2_arr = np.array([dummy_val(r) for r in res])

filter_inds = np.not_equal(test2_arr, res2_arr)
incorrectly_clf_X = test_X[filter_inds]
incorrectly_clf_y = test_y[filter_inds]
```

```python
for i, text in enumerate(incorrectly_clf_X):
    print(f"Class: {incorrectly_clf_y.iloc[i]}, index: {incorrectly_clf_X.index[i]}")
    print(text)
    print()
```
