```python
import wordninja
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

def clean_text(text: str, stem: bool = True):
    text = text.lower()   
    remove_string = 'tcb-script]'  # random string appearing in some text samples
    text = text.replace(remove_string, '')
    
    tokens = separate_words(text)
    
    # remove punctuation from each token
    table = str.maketrans('', '', string.punctuation)
    tokens = [w.translate(table) for w in tokens]
    
    tokens = [word for word in tokens if word.isalpha()]

    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words]
    
    if stem:
        porter = PorterStemmer()
        tokens = [porter.stem(word) for word in tokens]
    
    return tokens
    
def separate_words(text):
    return wordninja.split(text)
```

```python
import matplotlib

from functools import reduce
from collections import Counter
%matplotlib notebook

def plot_hist(text_list, n_top=30, title=""):
    word_counts = Counter(text_list)
    df = pandas.DataFrame.from_dict(word_counts, orient='index')
    df.columns = ['Word frequency']
    
    df = df.nlargest(n_top, 'Word frequency')
    df.plot(kind='bar', title=title)

ana_list = reduce(lambda a,b: a + b, ana_text_cleaned, [])
ok_list = reduce(lambda a,b: a + b, ok_text_cleaned, [])

plot_hist(ana_list, title="Pro-ana")
plot_hist(ok_list, title="Pro-recovery")

print(f"Ana unique words: {len(set(ana_list))}")
print(f"Ok unique words: {len(set(ok_list))}")
```

```python
from functools import partial
from sklearn.model_selection import train_test_split

#cleaned_data = data.copy()
cleaned_data['text'] = cleaned_data['text'].apply(partial(clean_text, stem=False))
cleaned_data['text'].head()
```

```python
with open('invalid_words.txt', 'r') as f:
    invalid_words = f.read().splitlines()    
print(invalid_words[:5])

def remove_invalid(text):
    return [w for w in text if w not in invalid_words]

cleaned_data['text'] = cleaned_data['text'].apply(remove_invalid)
```

```python
max_seq_len = cleaned_data['text'].apply(lambda t: len(t)).max()
max_seq_len
```

```python
cleaned_data['text'] = cleaned_data['text'].apply(lambda x: " ".join(x))
print(cleaned_data.head())

train, test = train_test_split(cleaned_data, test_size=0.1, random_state=42)
train_X, train_y = train['text'], train['label']
test_X, test_y = test['text'], test['label']

# cleaned_data.to_csv('cleaned_data.csv')
```

```python
cleaned_data['text']
```

```python
import tensorflow as tf
tf.test.is_gpu_available()
```

```python
from word2vec_keras import Word2VecKeras
model = Word2VecKeras()
model.train(train_X.to_list(),
            train_y,
            w2v_size=50,  # much smaller n words
            w2v_epochs=100,
            k_batch_size=32,
            k_max_sequence_len=max_seq_len,
            k_epochs=150,
            k_lstm_neurons=20)
```

```python
res = []

for text, label in zip(test_X, test_y):
    try:
        pred = model.predict(text)
        res.append(pred)
        print(f"{pred}, orig: {label}")
    except IndexError as e:
        print(e)
        res.append(None)
```

```python
from sklearn.metrics import accuracy_score

none_count = 0
res2 = []
test2 = []

for r, t in zip(res, test_y):
    if r is not None:
        res2.append(r['label'])
        test2.append(t)
    else:
        none_count += 1
        
print(none_count)
accuracy_score(test2, res2)
```

```python
import numpy as np

def dummy_val(r):
    return r['label'] if r is not None else -1

test2_arr = np.array(test_y)
res2_arr = np.array([dummy_val(r) for r in res])

filter_inds = np.not_equal(test2_arr, res2_arr)
incorrectly_clf_X = test_X[filter_inds]
incorrectly_clf_y = test_y[filter_inds]
```
